## Demonstration

```python
from arcgis.geoenrichment import get_countries, Country
```

```python
gis = gis_local
```

```python
get_countries(gis)
```

```python
cntry = Country('usa', gis=gis)

cntry
```

```python
ev = cntry.enrich_variables

ev
```

```python
kv = ev[
    (ev.data_collection.str.lower().str.contains('key'))
    & (ev.name.str.lower().str.contains('cy'))
].reset_index(drop=True)

kv
```

```python
in_df = demo_data.pdx_coffee_study_areas_3min.df
```

```python
edf = cntry.enrich(in_df, enrich_variables=kv)

edf.info()
edf.head()
```
