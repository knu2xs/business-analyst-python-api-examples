# Demonstration

## Demonstration Data

```python
study_area_pth = r'C:\projects\business-analyst-python-api-examples\data\interim\interim.gdb\pdx_coffee_study_areas_demo'
```

```python
from arcgis.features import GeoAccessor

df = GeoAccessor.from_featureclass(study_area_pth)

df.info()
df.head()
```

## GIS Source

```python
from arcgis.gis import GIS

gis = GIS('pro')
```

## Introspection - Available Countries

```python
from arcgis.geoenrichment import get_countries

get_countries(gis)
```

## Country Object Instance

```python
from arcgis.geoenrichment import Country

usa = Country('usa', gis=gis)

usa
```

## Introspction - Available Enrichment Variables

```python
ev = usa.enrich_variables

ev
```

### Select Variables for Analysis

```python
kv = ev.loc[
    (ev.name.str.lower().str.contains('cy'))
    & (ev.data_collection.str.lower().str.contains('key'))
].reset_index(drop=True)

kv
```

## Enrich

```python
enrich_df = usa.enrich(df, enrich_variables=kv)

enrich_df
```
