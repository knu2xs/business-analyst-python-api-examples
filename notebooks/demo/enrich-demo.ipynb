```python
from arcgis.features import GeoAccessor
from arcgis.gis import GIS

from arcgis.geoenrichment import get_countries, Country
```

```python
sa_pth = r'D:\projects\business-analyst-python-api-examples\data\raw\raw.gdb\pdx_coffee_study_areas_3min_small'
```

```python
df = GeoAccessor.from_featureclass(sa_pth)

df.info()
df.head()
```

```python
gis = GIS('pro')
```

```python
get_countries(gis)
```

```python
usa = Country('usa', gis=gis)

usa
```

```python
ev = usa.enrich_variables

ev
```

```python
kv = ev[
    (ev.name.str.lower().str.contains('cy'))
    & (ev.data_collection.str.lower().str.startswith('key'))
].reset_index(drop=True)

kv
```

```python
e_df = usa.enrich(df, kv)

e_df.info()
e_df.head()
```
