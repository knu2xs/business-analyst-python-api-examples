```python
fc_pth = r'D:\projects\business-analyst-python-api-examples\data\raw\raw.gdb\pdx_coffee_study_areas_3min'
```

```python
from arcgis.gis import GIS
```

```python
gis_pro = GIS('pro')
gis_agol = GIS(profile='ba')
```

```python
from arcgis.geoenrichment import get_countries, Country
```

```python
get_countries(gis_pro)
```

```python
usa = Country('usa', gis=gis_pro)

usa
```

```python
ev = usa.enrich_variables

ev
```

```python
kv = ev[
    (ev.name.str.lower().str.contains('cy'))
    & (ev.data_collection.str.lower().str.contains('key'))
].reset_index(drop=True)

kv
```

```python
from arcgis.features import GeoAccessor
```

```python
sa_df = GeoAccessor.from_featureclass(fc_pth)

sa_df.info()
sa_df.head()
```

```python
enrich_df = usa.enrich(sa_df, enrich_variables=kv)

enrich_df.info()
enrich_df.head()
```
