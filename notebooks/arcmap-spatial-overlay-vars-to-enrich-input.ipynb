```python
var_lst = var_str.replace("'", "").strip().split(';')

var_lst
```

```python
nm_lst = [var.split('.')[0] for var in var_lst]

nm_lst
```

## Look Up the Variables from the List

```python
usa = Country('usa')

usa
```

```python
var_df = usa._ba_cntry.get_enrich_variables_from_name_list(nm_lst)

var_df
```

## Get the Semicolon Delimited String for Input to the Enrich Layer Tool

```python
';'.join(var_df.enrich_name.values)
```

## Get Match Table

```python
input_df = pd.DataFrame(var_lst, columns=['input_string'])

input_df.head()
```

```python
input_df['name'] = input_df.input_string.str.split('.').apply(lambda var: var[0])

input_df.head()
```

```python
input_df.set_index('name').join(var_df.set_index('name').loc[:,['enrich_name']])
```
